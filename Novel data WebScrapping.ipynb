```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
abc=requests.get('https://www.goodreads.com/list/show/39291.Daily_Telegraph_500_Books_That_Everyone_Should_Read',headers=headers).text
```

```python
abc
```

```python
soup=BeautifulSoup(abc,'lxml')
```

```python
soup.Prettify
```

```python
soup
```

```python
xyz=soup.find('div',class_='leftContainer')
```

```python
xyz
```

```python
name=[]
for i in xyz.find_all('a',class_='bookTitle'):
    name.append(i.text.strip())
```

```python
name
```

```python
author=[]
for i in xyz.find_all('a',class_='authorName'):
    author.append(i.text.strip())
```

```python
author
```

```python
len(author)
```

```python
authore=[]
for i in xyz.find_all('span',class_='minirating'):
    authore.append(i.text.strip())
```

```python
import re

# Initialize an empty list to store the extracted floats
extracted_floats = []

# Define a regex pattern to match floats
pattern = r"[-+]?\d*\.\d+|\d+"

# Loop through each string in the list
for text in authore:
    # Use findall to extract all floats from the current string
    floats = re.findall(pattern, text)
    
    # Check if floats were found in the current string
    if floats:
        # Convert the first float to a float type and append it to the result list
        first_float = float(floats[0])
        extracted_floats.append(first_float)

# Print the extracted floats
print(extracted_floats)
```

```python
len(extracted_floats)
```

```python
d={'Name of Novel':name,'Author':author,'Ratings':extracted_floats}
```

```python
df=pd.DataFrame(d)
```
